This is an attempt at a classification model for xG

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import atan2, degrees, dist

from statsbombpy import sb
from mplsoccer import Pitch, VerticalPitch

from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

## Collect data

```python
# collect all passing events 
shot_df = pd.DataFrame(columns=['teammate', 'actor', 'location_x', 'id',
           'player', 'position', 'team', 'type', 'timestamp', 
           'under_pressure', 'goalkeeper_position', 'shot_outcome',
           'shot_statsbomb_xg', 'shot_freeze_frame'])

matches = sb.matches(competition_id=55, season_id=43)
match_ids = matches[matches['match_status_360'] == 'available'].match_id.unique()

for match_id in tqdm(match_ids):
    
    events = sb.events(match_id=match_id)
    match_frames = sb.frames(match_id=match_id, fmt='dataframe')

    # join the events to the frames 
    df = pd.merge(match_frames, events, left_on='id', right_on='id', how='left')
    
    # take relevant rows 
    df = df[['teammate', 'actor', 'location_x', 'id',
           'player', 'position', 'team', 'type', 'timestamp', 
           'under_pressure', 'goalkeeper_position', 'shot_outcome',
           'shot_statsbomb_xg', 'shot_freeze_frame']]
    
    df = df[df['type'] == 'Shot']
    df = df[df['actor'] == True]
    

    # append to overall df
    shot_df = shot_df.append(df)


len(shot_df.id.unique())
```

# Features

given this model will be used to assess the potential xG of a resultant cross, relevenat features would be:
- location
- goalkeeper_position
- shot_freeze_frame 
- outcome (Target - either 1 (goal) or 0 (OW))

Some features to be extrapolated / calculated:
- number of players in shot takers vacinity 
- number of players between shot taker and goal 
- distance of the shot taker to the net
- angle from middle of goal 

From research, Statsbomb has concluded that the radius of pressure is roughly 2.3 yards ~= 3 yards.

"That means the pressure radius is on average only, 2.3 yards"

https://statsbomb.com/articles/soccer/shots-under-pressure-part-2-headers/

## Generating additional features

```python
# https://www.geeksforgeeks.org/check-whether-a-given-point-lies-inside-a-triangle-or-not/
def area(x1, y1, x2, y2, x3, y3):
    return abs((x1*(y2-y3) + x2*(y3-y1)+ x3*(y1-y2))/2.0)

def in_net_cone(x1, y1, x2, y2, x3, y3, x, y):
    A = area (x1, y1, x2, y2, x3, y3)
    A1 = area (x, y, x2, y2, x3, y3)
    A2 = area (x1, y1, x, y, x3, y3)
    A3 = area (x1, y1, x2, y2, x, y)
    return A == A1 + A2 + A3
```

```python
def get_angle(x1, y1): #These can also be four parameters instead of two arrays
    angle = atan2(x1 - 120, y1 - 40)
    angle = degrees(angle)
    return round(abs(angle),2)
```

```python
loc_x, loc_y, gk_x, gk_y, num_opponents_around, num_opponents_bw_net, shot_dist, outcome = [],[],[],[],[],[],[],[]

for id, row in shot_df.iterrows():

    shot_x = row.loc_x
    shot_y = row.loc_y
    
    if 60-shot_y > 0:
        net_coords = [0,30,0,50]
    else:
        net_coords = [120,30,120,50]
    
    count_num_opponents_around = 0
    count_num_opponents_bw_net = 0
#     angle = get_angle(shot_x, shot_y)
    dist_to_net = dist([shot_x,shot_y], [net_coords[0], 40])
    shot_outcome = row.shot_outcome
    
    if shot_outcome == 'Goal': shot_outcome = 1
    else: shot_outcome = 0
    
    if not isinstance(row.shot_freeze_frame, float):
        for player in row.shot_freeze_frame:
            player_x = player['location'][0]
            player_y = player['location'][1]

            #count how many opponents were within 3 yards of player when shot was taken
            if ((shot_x - player_x)**2 + (shot_y - player_y)**2) <= 3**2 and (player['teammate'] == False) and (player['position']['name'] != 'Goalkeeper'):
                count_num_opponents_around += 1

            #count how many opponents were between shot and net
            if (player['teammate'] == False) and (player['position']['name'] != 'Goalkeeper'):
                if in_net_cone(net_coords[0], net_coords[1], net_coords[2], net_coords[3], shot_x, shot_y, player_x, player_y):
                    count_num_opponents_bw_net += 1

            #get position of opponent's goalkeeper 
            if ((player['position']['name'] == 'Goalkeeper') and (player['teammate'] == False)):
                gk_pos_x = player['location'][0]
                gk_pos_y = player['location'][1]
                
    loc_x.append(shot_x)
    loc_y.append(shot_y)
    gk_x.append(gk_pos_x)
    gk_y.append(gk_pos_y)
    num_opponents_around.append(count_num_opponents_around) 
    num_opponents_bw_net.append(count_num_opponents_bw_net) 
    shot_dist.append(dist_to_net) 
#     shot_angle.append(angle)
    outcome.append(shot_outcome)
    
    
d = {'loc_x': loc_x, 'loc_y': loc_y, 'gk_x': gk_x, 'gk_y': gk_y,
    'num_opponents_around': num_opponents_around, 'num_opponents_bw_net':num_opponents_bw_net,
    'shot_dist': shot_dist, 'outcome': outcome}
df = pd.DataFrame(data=d)
    
```

## Modelling

shot angle was an insignificant feature in this model as seen later in the notebook 

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import precision_score, recall_score, accuracy_score

# from xgboost import XGBClassifier
import xgboost as xgb

import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
X = df[['gk_x', 'gk_y','num_opponents_around',
        'num_opponents_bw_net', 'shot_dist']]
y = df['outcome']
```

```python
# train test split

train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=0.2,random_state=45,stratify=y)
```

### Address the imbalance in the target variable

```python
# oversampling
from collections import Counter
print(f"Training target statistics: {Counter(train_y)}")
print(f"Testing target statistics: {Counter(test_y)}")
```

```python
from imblearn.over_sampling import RandomOverSampler

over_sampler = RandomOverSampler(random_state=42)
X_res, y_res = over_sampler.fit_resample(train_X, train_y)

print(f"Training target statistics: {Counter(y_res)}")
```

# CV

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

model = LogisticRegression()
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l2', 'l1']
c_values = [1000, 100, 50, 10, 1.0, 0.1, 0.01]

grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='precision',error_score=0)
grid_result = grid_search.fit(X_res, y_res)

print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

## Fit cv model

```python
model = LogisticRegression(**grid_result.best_params_)
model.fit(X_res, y_res)
y_pred = model.predict(test_X)
accuracy = accuracy_score(test_y, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
print('******** LR **********')
print('Precision Score: ',precision_score(test_y,y_pred))
print('Recall Score: ',recall_score(test_y,y_pred))
print('Accuracy Score: ',accuracy_score(test_y,y_pred))
print('***********************')
```

```python
cm=confusion_matrix(test_y, y_pred)
sns.heatmap(cm,cmap='BuPu',annot=True,fmt='d', xticklabels=['not a goal', 'goal'], yticklabels=['not a goal', 'goal'])
```
